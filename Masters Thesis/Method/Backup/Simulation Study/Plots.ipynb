```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r'C:\Users\archi\OneDrive\Desktop\VU\My VU\Thesis\Method\Simulation Study\Results\Delta_Results_All.xlsx')
```

```python
df_300 = df[df['T'] == 300]
df_500 = df[df['T'] == 500]
df_800 = df[df['T'] == 800]
```

### 300

```python
df = df_300.reset_index()
delta_col = "delta"
num_series_col = "num_series"
sadf_col = "SADF"
crit_95_col = "crit_95"

group_25 = df[df['num_series'] == 25][['delta', 'SADF']].rename(columns={'SADF': 'SADF_25'})
group_15  = df[df['num_series'] == 15][['delta', 'SADF']].rename(columns={'SADF': 'SADF_15'})
group_5  = df[df['num_series'] == 5][['delta', 'SADF']].rename(columns={'SADF': 'SADF_5'})
group_2  = df[df['num_series'] == 2][['delta', 'SADF']].rename(columns={'SADF': 'SADF_2'})

# Merge on 'delta'
merged = group_25.merge(group_5, on='delta', how='outer').merge(group_2, on='delta', how='outer').merge(group_15, on='delta', how='outer')

plt.figure(figsize=(10, 6))

plt.plot(merged['delta'], merged['SADF_25'],marker = 'o', label='num_series = 25', color='tab:blue')
plt.plot(merged['delta'], merged['SADF_15'],marker = 'o', label='num_series = 15', color='tab:purple')
plt.plot(merged['delta'], merged['SADF_5'],marker = 'o', label='num_series = 5', color='tab:orange')
plt.plot(merged['delta'], merged['SADF_2'],marker = 'o', label='num_series = 2', color='tab:green')

# Critical value lines
plt.axhline(y=2.435, color='tab:blue', linestyle='--', label='crit_95 (25)')
plt.axhline(y=2.441, color='tab:purple', linestyle='--', label='crit_95 (15)', alpha = 0.7)
plt.axhline(y=2.01, color='tab:orange', linestyle='--', label='crit_95 (5)')
plt.axhline(y=2.287, color='tab:green', linestyle='--', label='crit_95 (2)')
plt.xticks(merged['delta'], rotation=45)

plt.xlabel('Delta')
plt.ylabel('SADF')
plt.title('SADF for each delta for different num_series (N = 300)')
plt.legend(loc='upper left')
plt.grid(True,axis='x')
plt.tight_layout()
plt.savefig(r"C:\Users\archi\OneDrive\Desktop\VU\My VU\Thesis\Method\Simulation Study\Plots\300_delta.png", dpi=300, bbox_inches='tight')
plt.show()
```

##### this tell us 2 things - as delta increases for lower number of series 0.01 is a spot after which the SADF increases exponentially
##### for lower num_series, it takes time for SADF to increase as 2 --> 5 but then there is a spot after which the initial SADF decreases for smaller delta

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1. Load your data
df = pd.read_excel(r"C:\Users\archi\OneDrive\Desktop\VU\My VU\Thesis\Method\Simulation Study\Results\Results_All.xlsx", sheet_name="Sheet1")

# 2. Choose your combination   
rw = 0.6   
for ns in [2,5,10,25,50,100]:
    # 3. Filter
    subset = df[(df['num_series']==ns) & (df['num_of_rw']==rw)]
    
    # 4. Prepare xaxis positions
    Ts = sorted(subset['T'].unique())
    x  = np.arange(len(Ts))
    bar_width = 0.1
    
    # 5. Define your palettes
    green_palette = ['#bfcf1e','#6cbe07','#339f03']   # for rw=0.4
    warm_palette  = ['#e8a001','#e96a1e','#ea364d']   # for rw=0.6
    palette = warm_palette if rw==0.6 else green_palette
    
    # 6. Plot each coint_frac
    coint_fracs = sorted(subset['coint_frac'].unique())
    plt.figure(figsize=(8,5))
    plt.grid()
    
    for i, frac in enumerate(coint_fracs):
        # collect SADF in the same Torder
        sadf_vals = [
            subset.loc[(subset['T']==t)&(subset['coint_frac']==frac),'SADF'].iloc[0]
            for t in Ts
        ]
        plt.bar(x + i*bar_width, sadf_vals,
                width=bar_width,
                color=palette[i],
                label=f'coint_frac={frac}')
    
    # 7. Ticks & labels
    plt.xticks(x + bar_width, Ts)
    plt.xlabel("T")
    plt.ylabel("SADF")
    plt.title(f"SADF for num_series={ns}, num_of_rw={rw}")
    plt.legend()
    plt.tight_layout()
    plt.savefig(r"C:\Users\archi\OneDrive\Desktop\VU\My VU\Thesis\Method\Simulation Study\Plots\SADF_main_" + str(rw)+"_"+str(ns)+".png", dpi=300, bbox_inches='tight')
    plt.show()
```
