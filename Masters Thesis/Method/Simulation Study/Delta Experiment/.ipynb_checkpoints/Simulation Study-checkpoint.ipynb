```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
import statsmodels.api as sm
from statsmodels.tsa.stattools import coint
np.random.seed(42)
```

#### Functions

```python
def random_walk(T):
    e = np.random.normal(0,1,T)
    z = np.cumsum(e)
    return z

def adf_fixed_lag0(series):
    return adfuller(series, maxlag=0, regression="c", autolag=None)[0]

def sadf(series, min_window):
    T = len(series)
    stats = []
    for end in range(min_window, T + 1):
        sub_series = series[:end]
        stat = adf_fixed_lag0(sub_series)
        stats.append(stat)
    return max(stats)

def sadf_critical_values_from_simulation(T, B=1000, seed=None):
    if seed is not None:
        np.random.seed(seed)
    r0 = 0.01 + 1.8 / np.sqrt(T)
    min_window = int(np.floor(r0 * T))
    stats = []
    for _ in range(B):
        series = random_walk(T)
        stat = sadf(series, min_window)
        stats.append(stat)
    stats = np.sort(stats)
    return {
        "90%": np.percentile(stats, 90),
        "95%": np.percentile(stats, 95),
        "99%": np.percentile(stats, 99),
        "all": stats
    }
    
def generate_bubble_series_from_z(z, r, beta=1.0, delta=0.05, sigma_eps=1.0, seed=None):
    if seed is not None:
        np.random.seed(seed)

    T = len(z)
    tau1 = int(np.floor(r * T))
    
    y = np.zeros(T)
    eps = np.random.normal(0, sigma_eps, T)

    for t in range(1, T):
        if t <= tau1:
            y[t] = beta * z[t] + eps[t]
        else:
            y[t] = (1 + delta) * y[t-1] + eps[t]
    
    return y

def sadf_series(series, min_window):
    T = len(series)
    stats = []
    for end in range(min_window, T + 1):
        sub_series = series[:end]
        stat = adf_fixed_lag0(sub_series)
        stats.append(stat)
    return stats
```

### Generating Critical Values

```python
T_values = [100, 300, 500, 1000]
results = {}
crit_table = []

for T in T_values:
    print("Running for T:", T)
    res = sadf_critical_values_from_simulation(T, B=1000, seed=42)
    results[T] = res
    crit_table.append({
        "T": T,
        "90%": round(res["90%"], 3),
        "95%": round(res["95%"], 3),
        "99%": round(res["99%"], 3)
    })
    
    plt.figure(figsize=(6,4))
    plt.hist(res['all'], bins=40, density=True, alpha=0.7, color='steelblue')
    plt.axvline(res['90%'], color='orange', linestyle='--', label='90%')
    plt.axvline(res['95%'], color='red', linestyle='--', label='95%')
    plt.axvline(res['99%'], color='purple', linestyle='--', label='99%')
    plt.title(f"SADF Null Distribution (T={T})")
    plt.xlabel("SADF Statistic")
    plt.ylabel("Density")
    plt.legend()
    plt.tight_layout()
    plt.show()

df = pd.DataFrame(crit_table)

fig, ax = plt.subplots(figsize=(6, 1.5))
ax.axis('off')
table = ax.table(cellText=df.values, colLabels=df.columns, loc='center', cellLoc='center')
table.scale(1, 1.5)
plt.title("SADF Critical Values Table", pad=20)
plt.show()
```
