```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
import statsmodels.api as sm
from statsmodels.tsa.stattools import coint
import my_functions as mf
np.random.seed(42)
```

```python
import sys
def loading_bar(progress, total, length=30):
    percent = int(100 * progress / total)
    filled = int(length * progress / total)
    bar = 'â–ˆ' * filled + '-' * (length - filled)
    sys.stdout.write(f'\rRunning Simulations: |{bar}| {percent}%')
    sys.stdout.flush()
```

### T = 100

```python
r_values = [0.2, 0.4, 0.5, 0.6, 0.8]
simulations = 1000
T = 100
beta = np.array([0.2, 0.5])
z = mf.random_walk(T)

summary_stats = []
plt.figure(figsize=(15, 10))

for idx, r in enumerate(r_values):
    sadf_values = []
    
    for _ in range(simulations):
        eps = np.random.normal(0, 1, T)
        y2 = mf.generate_bubble_series_from_z(z=z, r=r, beta=beta[1], delta=0.05)
        
        r0 = 0.01 + 1.8 / np.sqrt(T)
        min_window = int(np.floor(r0 * T))
        sadf_path = mf.sadf_series(y2, min_window)
        sadf_stat = max(sadf_path)
        sadf_values.append(sadf_stat)
    
    summary_stats.append({
        'r': r,
        'min_SADF': min(sadf_values),
        'max_SADF': max(sadf_values)
    })
    
    plt.subplot(3, 2, idx+1)
    plt.hist(sadf_values, bins=40, density=True, alpha=0.6, color='blue')
    plt.title(f"SADF Distribution (r = {r}) & T = {T}")
    plt.xlabel("SADF")
    plt.ylabel("Density")

plt.tight_layout()
plt.show()

df_summary = pd.DataFrame(summary_stats).set_index('r')
print(df_summary)

```
