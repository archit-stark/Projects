```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r'C:\Users\archi\OneDrive\Desktop\VU\My VU\Thesis\Method\Simulation Study\Results\Delta_Results_All.xlsx')
```

```python
df_300 = df[df['T'] == 300]
df_500 = df[df['T'] == 500]
df_800 = df[df['T'] == 800]
```

### 300

```python
df = df_300.reset_index()
delta_col = "delta"
num_series_col = "num_series"
sadf_col = "SADF"
crit_95_col = "crit_95"

group_25 = df[df['num_series'] == 25][['delta', 'SADF_mean']].rename(columns={'SADF_mean': 'SADF_25'})
group_15  = df[df['num_series'] == 15][['delta', 'SADF_mean']].rename(columns={'SADF_mean': 'SADF_15'})
group_5  = df[df['num_series'] == 5][['delta', 'SADF_mean']].rename(columns={'SADF_mean': 'SADF_5'})
group_2  = df[df['num_series'] == 2][['delta', 'SADF_mean']].rename(columns={'SADF_mean': 'SADF_2'})

# Merge on 'delta'
merged = group_25.merge(group_5, on='delta', how='outer').merge(group_2, on='delta', how='outer').merge(group_15, on='delta', how='outer')

plt.figure(figsize=(10, 6))

plt.plot(merged['delta'], merged['SADF_25'],marker = 'o', label='num_series = 25', color='#07b9f0')
plt.plot(merged['delta'], merged['SADF_15'],marker = 'o', label='num_series = 15', color='#1361ad')
plt.plot(merged['delta'], merged['SADF_5'],marker = 'o', label='num_series = 5', color='#a2a3a3')
plt.plot(merged['delta'], merged['SADF_2'],marker = 'o', label='num_series = 2', color='#434545')

# Critical value lines
plt.axhline(y=2.795, color='#07b9f0', linestyle='--', label='crit_95 (25)')
plt.axhline(y=2.392, color='#1361ad', linestyle='--', label='crit_95 (15)', alpha = 0.7)
plt.axhline(y=3.027, color='#a2a3a3', linestyle='--', label='crit_95 (5)')
plt.axhline(y=3.201, color='#434545', linestyle='--', label='crit_95 (2)')
plt.xticks(merged['delta'], rotation=45)

plt.xlabel('Delta')
plt.ylabel('SADF (mean)')
plt.title('SADF for each delta for different num_series (N = 300)')
plt.legend(loc='upper left')
plt.grid(True,axis='x')
plt.tight_layout()
plt.savefig(r"C:\Users\archi\OneDrive\Desktop\VU\My VU\Thesis\Method\Simulation Study\Plots\Delta Plots\300_delta_SADF.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1. Load your data
df = pd.read_excel(r"C:\Users\archi\OneDrive\Desktop\VU\My VU\Thesis\Method\Simulation Study\Results\Results_All.xlsx", sheet_name="Sheet1")

# 2. Choose your combination   
rw = 0.6   
for ns in [2,5,10,25,50,100]:
    # 3. Filter
    subset = df[(df['num_series']==ns) & (df['num_of_rw']==rw)]
    
    # 4. Prepare xaxis positions
    Ts = sorted(subset['T'].unique())
    x  = np.arange(len(Ts))
    bar_width = 0.125
    
    # 5. Define your palettes
    green_palette = ['#bfcf1e','#6cbe07','#339f03']   # for rw=0.4
    warm_palette  = ['#9a9a9a', '#707070', '#454545']   # for rw=0.6
    palette = warm_palette if rw==0.6 else green_palette
    
    # 6. Plot each coint_frac
    coint_fracs = sorted(subset['coint_frac'].unique())
    plt.figure(figsize=(8,5))
    plt.grid(linewidth=0.5, alpha=0.6)
    
    for i, frac in enumerate(coint_fracs):
        # collect SADF in the same Torder
        sadf_vals = [
            subset.loc[(subset['T']==t)&(subset['coint_frac']==frac),'SADF_mean'].iloc[0]
            for t in Ts
        ]
        plt.bar(x + i*bar_width, sadf_vals,
                width=bar_width,
                color=palette[i],
                label=f'coint_frac={frac}')
    
    # 7. Ticks & labels
    plt.xticks(x + bar_width, Ts)
    plt.xlabel("T")
    plt.ylabel("SADF_mean")
    plt.title(f"SADF for num_series={ns}, num_of_rw={rw}")
    plt.legend()
    plt.tight_layout()
    plt.savefig(r"C:\Users\archi\OneDrive\Desktop\VU\My VU\Thesis\Method\Simulation Study\Plots\SADF Main\SADF_main_" + str(rw)+"_"+str(ns)+".png", dpi=300, bbox_inches='tight')
    plt.show()
```

```python
coint_frac_values = [0.4, 0.6, 0.8]
num_series_values = [2, 5, 10, 25, 50, 100]
num_of_rw_values = [0.4, 0.6]

for T_val in sorted(df['T'].unique()):
    for num_series in num_series_values:
        for num_of_rw in num_of_rw_values:
            box_stats = []
            for coint_frac in coint_frac_values:
                subset = df[
                    (df['T'] == T_val) &
                    (df['num_series'] == num_series) &
                    (df['num_of_rw'] == num_of_rw) &
                    (df['coint_frac'] == coint_frac)
                ]
                if subset.empty:
                    continue
                row = subset.iloc[0]
                box_stats.append({
                    'label': str(int(coint_frac*100)),
                    'whislo': row['SADF_min'],
                    'q1':     row['SADF_p25'],
                    'med':    row['SADF_median'],
                    'q3':     row['SADF_p75'],
                    'whishi': row['SADF_max'],
                    'fliers': [row['SADF_p95'], row['SADF_p99']],
                })
            if box_stats:
                fig, ax = plt.subplots(figsize=(6, 4))
                ax.bxp(
                    box_stats,
                    patch_artist=True,  # < needed for box facecolor to apply
                    boxprops=dict(facecolor='#9a9a9a', edgecolor='#9a9a9a'),
                    whiskerprops=dict(color='#9a9a9a'),
                    capprops=dict(color='#9a9a9a'),
                    medianprops=dict(color='#29a7f0', linewidth=2.5),
                )
                ax.set_title(f'T = {T_val}, No of series = {num_series}, % of Exploding Series = {num_of_rw * 100:.0f}%')
                ax.set_xlabel('Fraction of time series length used for Cointegration (%)')
                ax.set_ylabel('SADF Statistic Values')
                plt.tight_layout()
                plt.savefig(r"C:\Users\archi\OneDrive\Desktop\VU\My VU\Thesis\Method\Simulation Study\Plots\Box Plots\BOX_main_" + str(T_val) + "_" + str(num_of_rw )+"_"+str(num_series)+".png", dpi=300, bbox_inches='tight')
                plt.show()
```

```python
# 2. Fix the proportion of exploding series you want to plot
rw = 0.6

# 3. Loop over each number of series value
for ns in [2, 5, 10, 25, 50, 100]:
    # 4. Filter down to the rows where num_series = ns and num_of_rw = rw
    subset = df[(df["num_series"] == ns) & (df["num_of_rw"] == rw)]
    if subset.empty:
        continue

    # 5. Build the xaxis as the sorted unique values of T
    Ts = sorted(subset["T"].unique())
    x = np.arange(len(Ts))

    # 6. Choose a color palette: graytones if rw == 0.6, greentones otherwise
    # green_palette = ['#29a7f0', '#1b7bcc', '#1361ad'] 
    # warm_palette  = ['#e8a001','#e96a1e','#ea364d']
    palette =  ['#1195fa', '#d9dadb','#5c5c5c']

    # 7. Grab the distinct cointegration fractions, sorted
    coint_fracs = sorted(subset["coint_frac"].unique())

    # 8. Start a new figure for this (ns, rw) combination
    plt.figure(figsize=(8, 5))
    plt.grid(which="both", linestyle="--", linewidth=0.5, alpha=0.5)

    # 9. For each cointegration fraction, plot power * 100 as a line
    for i, frac in enumerate(coint_fracs):
        # Extract power in the same order as Ts, then multiply by 100
        power_vals = [
            subset.loc[
                (subset["T"] == t) & (subset["coint_frac"] == frac),
                "power"
            ].iloc[0]
            * 100
            for t in Ts
        ]
        plt.plot(
            x,
            power_vals,
            marker="o",
            color=palette[i],
            label=f"coint_frac = {frac}"
        )

    # 10. Ticks & labels & legend
    plt.xticks(x, Ts)
    plt.xlabel("T")
    plt.ylabel("Power (%)")
    plt.title(f"Power vs. T  num_series={ns}, % of Exploding Series={rw*100:.0f}%")
    plt.legend(loc="upper left")
    plt.tight_layout()

    # 11. Save to disk (adjust the path if needed)
    plt.savefig(r"C:\Users\archi\OneDrive\Desktop\VU\My VU\Thesis\Method\Simulation Study\Plots\Power Plots\Power_main_" + str(rw)+"_"+str(ns)+".png", dpi=300, bbox_inches='tight')
    plt.show()
```
