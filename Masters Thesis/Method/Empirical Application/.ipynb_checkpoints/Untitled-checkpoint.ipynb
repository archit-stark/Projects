```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
import pandas as pd
import numpy as np
import my_functions as mf
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.vector_ar.vecm import coint_johansen
import sounddevice as sd
```

```python
num_series = 14
z = mf.random_walk(1000).reshape(-1, 1) #constant
eps = np.random.normal(0, 1, size=(1000, num_series)) # DEPENDS on T
et = np.random.normal(0, 1, size=(1000, num_series)) #constant

B = 500
delta = 0.01
multipliers = mf.unique_rounded_uniform(0, 2, num_series, decimals=2) #constant

T = 156
e = eps[:T,:].copy()
common_fac = z[:T].copy()

num_of_rw = 0.6
coint_frac = 0.6
innov_8 = np.random.normal(0, 1, size=(T - int(T * coint_frac), int(num_of_rw * num_series))) #constant

innov = innov_8
crit_val = mf.simulate_sadf_distribution(B=B, T=T, num_series=num_series, coint_frac=coint_frac, num_of_rw=num_of_rw, multi = multipliers)
print(crit_val)
crit_val_95 = np.round(crit_val.loc["95%", "Critical Value"],3)
crit_val_99 = np.round(crit_val.loc["99%", "Critical Value"],3)

print("\n")

""" Run Experiment """
mf.run_sadf_bubble_check( delta, T, coint_frac, num_of_rw,  common_fac, multipliers, innov, e, crit_val_95, crit_val_99, num_series)
print("Multipliers:", multipliers)
```

```python
df = pd.read_csv("Unemp_Rate.csv", index_col = ["TIME"])
```

```python
df
```

```python
plt.figure(figsize=(15, 5))
plt.plot(range(1, len(df) + 1), df)
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Time Series Plot")
plt.legend(df.columns)
plt.show()
```

```python

df = np.array(df)
Y = df[0:100, 0]
X = df[0:100, 1:]
X_const = sm.add_constant(X)
model = sm.OLS(Y, X_const)
results = model.fit()
params = results.params
a = params[0]  # Intercept
b = params[1:]
pre_resid = Y - (a + X@b)
```

```python
result = adfuller(pre_resid)
print(f'ADF statistic: {result[0]}, p-value: {result[1]}')

if result[1] > 0.05:
    print("Series is non-stationary. Try differencing.")
```

```python
plt.plot(pre_resid)
```

```python
Y = df[100:, 0]
X = df[100:, 1:]
post_resid =  Y - (a + X@b)
```

```python
resid = np.concatenate([pre_resid, post_resid])
```

```python
plt.plot(post_resid)
```

```python
T = len(resid)
r0 = 0.01 + 1.8 / np.sqrt(T)
min_window = int(np.floor(r0 * T))
sadf = mf.sadf_series(resid, min_window)
```

```python
T
```

```python

crit_95 = crit_val_95
crit_99 = crit_val_99

plt.plot(sadf, label='SADF Statistics')
#plt.axhline(crit_95, color='blue', linestyle='--', label='95% Critical Value')
plt.axhline(crit_99, color='red', linestyle='--', label='99% Critical Value')

plt.legend()
plt.xlabel("Time")
plt.ylabel("SADF Statistic")
plt.title("SADF Test with Critical Values")
plt.show()
```
