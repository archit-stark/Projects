### Question 1)

The first question of the assignment asks to explain the roles of \( \alpha_g \), \( \beta_g \), and \( \delta_g \) in the Google Ads adstock model:\\
$
gadst = \frac{\alpha_g \cdot g_t}{1 + \delta_g \cdot gadst_{-1}} + \beta_g \cdot gadst_{-1},]
$
and how they relate to the decay effect and diminishing returns.

1. **Role of \( \alpha_g \)**: This parameter represents the effectiveness of current advertising expenditure in generating adstock. Higher \( \alpha_g \) values indicate that each unit of expenditure has a stronger immediate impact on adstock. It directly scales the advertising input (\( g_t \)) and adjusts the initial level of consumer awareness or response to the ad.

2. **Role of \( \beta_g \)**: The \( \beta_g \) parameter captures the persistence of adstock over time, reflecting the "decay effect." A higher \( \beta_g \) implies a slower decay of the adstock, meaning the effects of past advertising linger longer. It determines how much of the previous period's adstock (\( gadst_{-1} \)) carries over into the current period.

3. **Role of \( \delta_g \)**: This parameter addresses the diminishing returns of advertising. As the previous adstock (\( gadst_{-1} \)) increases, the impact of additional advertising is reduced. A larger \( \delta_g \) value means a steeper reduction in marginal returns, indicating that higher past adstock levels make current advertising less effective in generating new awareness.

Together, these parameters allow the model to capture both the persistence of advertising effects (through \( \beta_g \)) and the diminishing impact of repeated advertising exposure (through \( \delta_g \)), while scaling the overall impact via \( \alpha_g \).

1. **Is the AB-testing procedure important?**
   - Yes, the AB-testing procedure is crucial in this context. AB-testing allows for a controlled experiment where one group (the "A" group) receives the existing advertising budget, while the other group (the "B" group) receives the increased budget (i.e., the impulse). This setup isolates the effect of the additional advertising expenditure, ensuring that the observed changes in sales can be attributed directly to the increased budget rather than to other factors that might simultaneously influence sales.

2. **Concerns about Simultaneity and Endogeneity Without AB-Testing**
   - **Simultaneity Issues:** In the absence of AB-testing, simultaneity can arise when advertising budgets and sales are determined simultaneously. For example, if the advertising budget is adjusted based on observed sales trends, this creates a two-way relationship where sales influence the advertising budget and the advertising budget influences sales. Such a scenario would introduce simultaneity bias, making it difficult to separate the cause-and-effect relationship.
   - **Endogeneity of Regressors:** Without AB-testing, the advertising budget (as an independent variable) may be endogenous. Endogeneity occurs when the independent variable is correlated with the error term in the regression model, often due to omitted variables, measurement errors, or simultaneity. In this case, if the decision to increase the advertising budget is related to unobserved factors that also affect sales, then the adstock variables (Google and YouTube) would be endogenous. This could bias the estimation of the coefficients \( \phi_g \) and \( \phi_y \), leading to unreliable results.

3. **Impact on the Interpretation of the Results**
   - **Bias in Estimated Coefficients:** If simultaneity and endogeneity are present, the estimated coefficients for the adstock variables (\( \phi_g \) and \( \phi_y \)) would be biased and inconsistent. This means that the estimated effect of an additional 100 euros spent on advertising would not accurately reflect the true causal impact on sales.
   - **Causal Inference Challenges:** The main goal of the impulse response analysis is to understand the causal effect of changes in the advertising budget on sales. Without AB-testing, any observed relationship between advertising expenditure and sales may reflect reverse causality or omitted variables, making it challenging to draw meaningful causal conclusions.

### Conclusion
The AB-testing procedure is essential for ensuring the validity of the causal interpretation in this case study. It helps to avoid simultaneity and endogeneity issues by isolating the effect of the advertising budget increase. Without it, the reliability of the modelâ€™s results would be compromised, leading to potentially misleading conclusions about the impact of advertising on sales.
