## Question II: Non-Linear Least Squares

```python
class LossFunctions:
    def __init__(self, data, params):
        self.data = data
        self.params = params

    def item_test(self, item, default):
        """
        Test if item is None, if it is return default
        """
        if item is not None:
            return item
        else:
            return default

    def ar1_loss(self, params=None, data=None):
        """
        AR(1) loss function
        """
        #Test if params and data are None
        params = self.item_test(params, self.params)
        data = self.item_test(data, self.data)

        #Initialize variables
        mu = params[0]
        delta = params[1]
        data = data

        #Calculate loss
        loss = 0
        for i in range(len(data) - 1):
            predicted = mu + delta * data['UNRATE_d'].iloc[i]
            loss += (data['UNRATE_d'].iloc[i+1] - predicted) ** 2   
        return loss
    
    def sestar_loss(self, params=None, data=None):
        """
        SESTAR loss function
        """
        #Test if params and data are None
        params = self.item_test(params, self.params)
        data = self.item_test(data, self.data)

        #Initialize variables
        data = self.data
        mu, delta, gamma, alpha, beta = params

        #Calculate loss
        loss = 0
        for i in range(len(data) - 1):
            logistic_term = 1 / (1 + np.exp(alpha + beta * data['UNRATE_d'].iloc[i]))
            predicted = mu + (delta + gamma * logistic_term) * data['UNRATE_d'].iloc[i]
            loss += (data['UNRATE_d'].iloc[i+1] - predicted) ** 2
        
        return loss
    
    def star_loss(self, params=None, data=None):
        """
        STAR loss function
        """
        #Test if params and data are None
        params = self.item_test(params, self.params)
        data = self.item_test(data, self.data)

        #Initialize variables
        data = self.data
        mu, delta, gamma, alpha, beta = params

        #Calculate loss       
        loss = 0
        for i in range(len(data) - 1):
            logistic_term = 1 / (1 + np.exp(alpha + beta * data['INDPRO_d'].iloc[i]))
            predicted = mu + (delta + gamma * logistic_term) * data['UNRATE_d'].iloc[i]
            loss += (data['UNRATE_d'].iloc[i+1] - predicted) ** 2
            
        return loss
```

```python
#Sanity check for loss functions
#Initial parameters
mu = 0
delta = 0.3
gamma = 1
alpha = 0
beta = 2
init_params = [mu, delta, gamma, alpha, beta]

loss_functions = LossFunctions(data, init_params) #Initialize loss functions class
#Calculate loss functions
ar1_loss = loss_functions.ar1_loss()
sestar_loss = loss_functions.sestar_loss()
star_loss = loss_functions.star_loss()

print(f"AR(1) loss: {round(ar1_loss,4)}")
print(f"SESTAR loss: {round(sestar_loss,4)}")
print(f"STAR loss: {round(star_loss,4)}")
```

## Question III: Non-Linear Least Squares Estimation

```python
class NLS_estimation:
    def __init__(self, data):
        self.data = data

    def print_results(self, params_est, model):
        """
        A function that prints estimated parameters
        """
        if model == "ar1":
            print(f"AR(1) model estimated parameters: mu = {round(params_est[0], 4)}, delta = {round(params_est[1], 4)}")
        elif model == "sestar":
            print(f"SESTAR model estimated parameters: mu = {round(params_est[0], 4)}, delta = {round(params_est[1], 4)}, gamma = {round(params_est[2], 4)}, alpha = {round(params_est[3], 4)}, beta = {round(params_est[4], 4)}")
        elif model == "star":
            print(f"STAR model estimated parameters: mu = {round(params_est[0], 4)}, delta = {round(params_est[1], 4)}, gamma = {round(params_est[2], 4)}, alpha = {round(params_est[3], 4)}, beta = {round(params_est[4], 4)}")
        else:
            print("Model not found")
    
    def ar1_est(self, params):
        """
        AR(1) estimation
        """
        data = self.data 
        loss_function = LossFunctions(self.data, params) #initialize loss function class
        result = minimize(loss_function.ar1_loss, params, args=(data,), method = 'BFGS') #minimize loss function for AR(1)
        mu, delta = result.x #extract estimated parameters
        params_est = [mu, delta]
        self.print_results(params_est, "ar1")
        return params_est
    
    def sestar_est(self, params):
        """
        SESTAR estimation
        """
        data = self.data
        loss_function = LossFunctions(self.data, params) #initialize loss function class
        result = minimize(loss_function.sestar_loss, params, args=(data,), method = 'BFGS') #minimize loss function for SESTAR
        mu, delta, gamma, alpha, beta = result.x #extract estimated parameters
        params_est = [mu, delta, gamma, alpha, beta]
        self.print_results(params_est, "sestar")
        return params_est
    
    def star_est(self, params):
        """
        STAR estimation
        """
        data = self.data
        loss_function = LossFunctions(self.data, params) #initialize loss function class
        result = minimize(loss_function.star_loss, params, args=(data,), method = 'BFGS') #minimize loss function for STAR
        mu, delta, gamma, alpha, beta = result.x #extract estimated parameters
        params_est = [mu, delta, gamma, alpha, beta]
        self.print_results(params_est, "star")
        return params_est
        
```

```python
non_linear_estimation = NLS_estimation(data) #initialize NLS_estimation class

#Estimate AR(1)
initial_param_ar1 = [0, 0.3]
ar_estimation = non_linear_estimation.ar1_est(initial_param_ar1)
print(ar_estimation)

#Estimate SESTAR
initial_param_sestar = [0, 0.3, 0, 0, 6]
sestar_estimation = non_linear_estimation.sestar_est(initial_param_sestar)
print(sestar_estimation)

#Estimate STAR
initial_param_star = [0, 0.3, 0, 0, 6]
star_estimation = non_linear_estimation.star_est(initial_param_star)
print(star_estimation)
```

```python
class FittedModels:
    def __init__(self, data):
        self.data = data

    def ar1_fitted(self, params):
        """
        Fit AR(1) model with estimated parameters
        """
        #initialize variables
        data = self.data
        mu, delta = params

        #Calculate fitted values
        fitted_data = []
        for i in range(len(data) - 1):
            predicted = mu + (delta * data['UNRATE_d'].iloc[i])
            fitted_data.append(predicted)
        return fitted_data
    
    def sestar_fitted(self, params):
        """
        Fit SESTAR model with estimated parameters
        """
        #initialize variables
        data = self.data
        mu, delta, gamma, alpha, beta = params

        #Calculate fitted values
        fitted_data = []
        for i in range(len(data) - 1):
            log_term = 1 / (1 + np.exp(alpha + beta * data['UNRATE_d'].iloc[i]))
            predicted = mu + (delta + gamma * log_term) * data['UNRATE_d'].iloc[i]
            fitted_data.append(predicted)
        return fitted_data
    
    def star_fitted(self, params):
        """
        Fit STAR model with estimated parameters
        """
        #initialize variables
        data = self.data
        mu, delta, gamma, alpha, beta = params

        #Calculate fitted values
        fitted_data = []
        for i in range(len(data) - 1):
            log_term = 1 / (1 + np.exp(alpha + beta * data['INDPRO_d'].iloc[i]))
            predicted = mu + (delta + gamma * log_term) * data['UNRATE_d'].iloc[i]
            fitted_data.append(predicted)
        return fitted_data
    
    def r_squared(self, model, params):
        """
        Calculate R-squared and adjusted R-squared of the fitted models
        """
        data = self.data

        #Model selection
        if model == 'AR1':
            fitted_data = self.ar1_fitted(params)
        elif model == 'SESTAR':
            fitted_data = self.sestar_fitted(params)
        elif model == 'STAR':
            fitted_data = self.star_fitted(params)

        #Calculate R-squared
        residuals = data['UNRATE_d'].iloc[1:] - fitted_data
        ss_res = np.sum(residuals ** 2)
        ss_tot = np.sum((data['UNRATE_d'].iloc[1:] - np.mean(data['UNRATE_d'].iloc[1:])) ** 2)
        r_squared = 1 - (ss_res / ss_tot)

        #Calculate adjusted R-squared
        n = len(data['UNRATE_d'].iloc[1:])  
        p = len(params)  
        r_squared_adj = 1 - ((1 - r_squared) * (n - 1) / (n - p - 1))
        return r_squared, r_squared_adj
```

```python
# Generate fitted values using the fitted functions
fitted_models = FittedModels(data)
fitted_ar1 = fitted_models.ar1_fitted(params=ar_estimation)
fitted_sestar = fitted_models.sestar_fitted(params=sestar_estimation)
fitted_star = fitted_models.star_fitted(params=star_estimation)
```

In examining the three periods of interestâ€”November 1973 to March 1975, January 1980 to July 1980, and December 2007 to June 2009â€”it becomes clear that each model responds differently to economic shocks.

The AR(1) model provides a basic linear relationship based on previous unemployment, leading to a relatively smooth fit. It fails to capture the sharp spikes in unemployment during the recessions, particularly in the Great Recession (2007-2009), where the model smooths over the volatility, underestimating the sudden shifts in the unemployment rate.

The SESTAR model introduces non-linearity, allowing for regime switching. This results in a better fit during periods of economic instability. For example, during the 1973-1975 and 1980 recessions, the SESTAR model captures some fluctuations, but it still struggles to fully reflect the magnitude of volatility during extreme periods.

The STAR model, with its more complex non-linear structure, fits the data the best, particularly during high-volatility periods like the Great Recession. It captures sudden changes in unemployment far better than both the AR(1) and SESTAR models, especially in 2008. However, this model may slightly overfit the data during more stable periods.

In conclusion, the AR(1) model provides a smoother fit, while SESTAR and STAR models progressively improve the fit during periods of significant economic stress.








## Question IV: Relationship Persistant Unemployment & Industrial Production

```python
# STAR Optimized Parameters
# Function to compute G(z_t-1, Î¸) for the STAR model
def G_star(INDPRO_d, star_params):
    delta = star_params[1]
    gamma = star_params[2]
    alpha = star_params[3]
    beta = star_params[4]

    G_values = []
    for z_t_1 in INDPRO_d:
        logistic_term = 1 / (1 + np.exp(alpha + beta * z_t_1))
        G = delta + gamma * logistic_term
        G_values.append(G)
    return G_values

# Calculate G(z_{t-1}, Î¸)
G_values = G_star(data['INDPRO_d'], star_estimation)

# Plot the values of INDPRO_d and G(z_{t-1}, Î¸) in a single line plot
plt.figure(figsize=(10, 4))

plt.plot(data.index, data['INDPRO_d'], label='Differenced INDPRO (INDPRO_d)', color='black')
plt.plot(data.index, G_values, label='Time-Varying Autoregressive Parameter $G(z_{t-1}$, Î¸)', color='red')

plt.title('Differenced Industrial Production and Time-Varying Parameter $G(z_{t-1}$, Î¸)')
plt.xlabel('Date')
plt.ylabel('Value')
plt.gca().set_facecolor('lightgray')
plt.legend()

plt.tight_layout()
plt.show()
```

The relationship between the persistence in unemployment rates and industrial production reflects the dynamics of economic cycles. Industrial production tends to be more volatile, reacting quickly to changes in economic conditions, such as demand, investment, and business cycles. On the other hand, unemployment rates typically adjust more slowly, reflecting persistence. This persistence is due to structural factors in the labor marketâ€”such as contracts, hiring and firing costs, and policy responsesâ€”that cause delays in employment adjustments to economic shocks.

Economic theory supports this observation. During a downturn, industrial production may decline sharply as firms reduce output in response to lower demand. However, firms may delay laying off workers due to costs or expectations of recovery, leading to a slower increase in unemployment. Similarly, during an expansion, production may rise quickly, but hiring lags behind as firms take time to assess the sustainability of the growth.

This persistence in unemployment is consistent with the time-varying autoregressive parameter \( G(z_{t-1}, \theta) \) seen in the STAR model plot. The smooth changes in \( G \) suggest that the economy transitions gradually between states, while industrial production experiences more frequent and sharper changes, aligning with observed economic cycles.

## Question VI: Robust Standard Errors

```python
class StandardErrors:
    def __init__(self, data):
        self.data = data

    def ar1_loss_function(self, params, hes='False'):
        """
        Calculate residuals for AR(1) model, based on estimated parameters
        and return residuals squared and the sum when used in Hessian (hes=True)
        """
        #initialize variables
        data = self.data
        mu = params[0]
        delta = params[1]
        
        #Calculate residuals
        fitted_models = FittedModels(data)
        fitted_ar1 = fitted_models.ar1_fitted(params)
        residuals = data['UNRATE_d'][1:] - fitted_ar1

        #return residuals outcome based on hes
        if hes is True:
            return np.sum(residuals**2)
        else:
            return residuals.values**2
        
    def sestar_loss_function(self, params, hes='False'):
        """
        Calculate residuals for SESTAR model, based on estimated parameters
        and return residuals squared and the sum when used in Hessian (hes=True)
        """
        #initialize variables
        data = self.data
        mu, delta, gamma, alpha, beta = params

        #Calculate residuals
        fitted_models = FittedModels(data)
        fitted_sestar = fitted_models.sestar_fitted(params)
        residuals = data['UNRATE_d'][1:] - fitted_sestar

        #return residuals outcome based on hes
        if hes is True:
            return np.sum(residuals**2)
        else:
            return residuals.values**2
    
    def star_loss_function(self, params, hes='False'):
        """
        Calculate residuals for STAR model, based on estimated parameters
        and return residuals squared and the sum when used in Hessian (hes=True)
        """
        #initialize variables
        data = self.data
        mu, delta, gamma, alpha, beta = params

        #Calculate residuals
        fitted_models = FittedModels(data)
        fitted_star = fitted_models.star_fitted(params)
        residuals = data['UNRATE_d'][1:] - fitted_star

        #return residuals outcome based on hes
        if hes is True:
            return np.sum(residuals**2)
        else:
            return residuals.values**2
        
    def jacobian_function(self, params, model):
        """
        Calculate the Jacobian matrix for the loss functions, 
        based on the selected model.
        This provides the first order derivatives of the loss function (also known as the gradient)
        """
        #select model
        if model == 'AR1':
            jacobian_func = nd.Jacobian(self.ar1_loss_function)
        elif model == 'SESTAR':
            jacobian_func = nd.Jacobian(self.sestar_loss_function)
        elif model == 'STAR':
            jacobian_func = nd.Jacobian(self.star_loss_function)
        #calculate jacobian
        jacobian = jacobian_func(params)
        return jacobian
    
    def newey_west(self, params, model):
        """
        Calculate the Newey-West covariance matrix for the loss functions,
        based on the selected model.
        """
        p = 12 # number of lags
        T = len(self.data) - 1 # number of observations
        gradients = self.jacobian_function(params, model) # calculate gradients
        sigma_0 = np.cov(gradients.T) # calculate covariance matrix

        # Initialize covariance matrix
        nw_covariance = sigma_0.copy()
        # Calculate the contribution of each lag
        for j in range(1, p + 1):
            sigma_j = np.zeros_like(sigma_0)
            for t in range(j+2, T):
                sigma_j_term = (1/T) * np.outer(gradients[t], gradients[t - j])
                sigma_j += sigma_j_term
            nw_covariance += (1 - j / (p + 1)) * (sigma_j + sigma_j.T) 
        return nw_covariance
    
    def hessian_function(self, params, model):
        """
        Calculate the Hessian matrix for the loss functions,
        based on the selected model.
        This provides the second order derivatives of the loss function
        """
        #select model
        if model == 'AR1':
            hessian_func = nd.Hessian(partial(self.ar1_loss_function, hes=True))
        elif model == 'SESTAR':
            hessian_func = nd.Hessian(partial(self.sestar_loss_function, hes=True))
        elif model == 'STAR':
            hessian_func = nd.Hessian(partial(self.star_loss_function, hes=True))
        #calculate hessian
        hessian_data = hessian_func(params)
        return hessian_data
    
    def omega_matrix(self, params, model):
        """
        Calculate the Omega matrix for the loss functions,
        based on the selected model.
        This is required when the model is misspecified.
        """
        hessian = self.hessian_function(params, model) # calculate hessian
        T = len(self.data) - 1 # number of observations
        omega = (1/T) * hessian # calculate omega matrix
        omega = np.linalg.inv(omega) # invert omega matrix
        return omega
    
    def standard_errors(self, params, model):
        """
        Calculate standard errors for the loss functions,
        based on the selected model.
        Using the Newey-West covariance matrix and the omega matrix for misspecification.
        """
        omega = self.omega_matrix(params, model) # calculate omega matrix
        nw = self.newey_west(params, model) # calculate Newey-West covariance matrix
        T = len(self.data) - 1 # number of observations
        cov_matrix = (1 / T) * omega @ nw @ omega.T # calculate covariance matrix
        se = np.sqrt(np.diag(cov_matrix)) # calculate standard errors
        return se, cov_matrix
```

```python
std_errors = StandardErrors(data) #initialize StandardErrors class

#Calculate standard errors for AR(1), SESTAR and STAR models
ar_se, ar_cov = std_errors.standard_errors(ar_estimation, 'AR1')
print('AR: ', ar_se)

star_se, star_cov = std_errors.standard_errors(star_estimation, 'STAR')
print('STAR: ', star_se)

sestar_se, sestar_cov = std_errors.standard_errors(sestar_estimation, 'SESTAR')
print('SESTAR: ', sestar_se)
```

## Question VII: Wald Test for Non-Linear Time-Dependence Effects in SESTAR and STAR

```python
def wald_test(params, cov_matrix):
    """
    Calculate the Wald test for the estimated parameters beta and gamma,
    to see whether the SESTAR and STAR models have non-linear time-dependence effects.
    """
    parameters = ['mu', 'delta', 'gamma', 'alpha', 'beta'] # parameters
    df_cov = pd.DataFrame(cov_matrix, columns=parameters, index=parameters) # covariance matrix
    time_dependent_variables = ['gamma', 'beta'] # time-dependent variables
    df_cov = df_cov[df_cov.index.isin(time_dependent_variables)] # select time-dependent variables
    df_cov = df_cov[time_dependent_variables] # select time-dependent variables
    params = np.array([params[2], params[4]]) # select beta and gamma
    wald_stat = params.T @ np.linalg.inv(df_cov[['gamma', 'beta']].values) @ params # calculate Wald statistic
    df = len(params) # degrees of freedom
    p_value = 1 - chi2.cdf(wald_stat, df) # calculate p-value
    return p_value
```

```python
#Calculate Wald test for SESTAR and STAR models
wald_test_sestar = wald_test(sestar_estimation, sestar_cov)
print('SESTAR Wald Test p-value: ', wald_test_sestar)

wald_test_star = wald_test(star_estimation, star_cov)
print('STAR Wald Test p-value: ', wald_test_star)
```
